```python
def create_name(f,l):
  first = f.capitalize()
  last = l.capitalize()
  return f"{first} {last}" # wht u want to print out

f_n = create_name("faiz","zaki")
print(f_n)
```

```python
def bmi1(w,h):
  return (round(w/(h/100)**2,2))
```

```python
def bmi_calculator(name,age,h,w):
  bmi = bmi1(w,h)
  if bmi < 18.5:
    result = "Under_weight"
  elif bmi<24.9:
    result = "Healthy Weight"
  elif bmi<29.9:
    result = "Over Weight"
  else:
    result = "Obesity"

  return f"Hi {name}!!\nwelcome to Nexpert Analytics(Your Health Expert)\nYour BMI is {bmi}\nYour BMI classification for the age {age} is {result}"

s = bmi_calculator("Faiz",24,170,70)
print(s)
```
