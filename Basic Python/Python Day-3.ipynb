```python
def bmi1(weight, height_cm):
    height_m = height_cm / 100
    return weight / (height_m ** 2)

def bmi_calculator(name, age, height, weight, unit):
    # Convert meters to cm if needed
    if unit == "meters":
        height = height * 100

    bmi = bmi1(weight, height)

    if bmi < 18.5:
        result = "Under Weight"
    elif bmi < 24.9:
        result = "Healthy Weight"
    elif bmi < 29.9:
        result = "Over Weight"
    else:
        result = "Obesity"

    return (
        f"Hi {name}!!\n"
        f"Welcome to Nexpert Analytics (Your Health Expert)\n"
        f"Your BMI is {bmi:.2f}\n"
        f"Your BMI classification for the age {age} is {result}"
    )
```

```python
import gradio as gr

with gr.Blocks(title="BMI Calculator App") as app:
    gr.Markdown("## 🧮 BMI Calculator – By Park JI Soo")

    name = gr.Textbox(label="Name")
    age = gr.Number(label="Age", precision=0)
    weight = gr.Number(label="Weight (kg)")
    height = gr.Number(label="Height")

    unit = gr.Radio(
        choices=["cm", "meters"],
        label="Height Unit",
        value="meters"
    )

    output = gr.Textbox(label="Result", lines=5)

    button = gr.Button("Calculate BMI")

    button.click(
        bmi_calculator,
        inputs=[name, age, height, weight, unit],
        outputs=output
    )

app.launch()
```

```python
def counter(a):
  a = a.replace(" ","")
  v_counter = 0
  for i in a:
    if i.lower() in "aeiou":
      v_counter += 1
  return f"The vowels count is {v_counter}. The consonants count is {len(a)-v_counter}"
```
