```python
#ATM simulator

#get total balance from user
#ask him to withdraw the cash till he reaches balance limit
#minimum balance limit RM 10

balance = int(input("Enter the total balance : "))
while balance>10:
  wd = int(input("Enter your withdraw :"))

  if balance - wd >= 10:
    balance = balance - wd
    print(f"Withdrawl of  RM {wd} is success")
    print(f"Remaining balance is RM {balance}")
  else:
    print("cannot withdraw")
    print("Minimum of RM 10 need to be maintained")
    break

print(f"Final balance is RM {balance}")


```

```python
#Login Attempts
attempts = 0
while attempts < 5:
    attempts += 1
    if attempts == 3:
        print("Account Locked")
        break
    print("Try Again")
```

```python
#ATM Pin Validation
pin = 1234
attempt = 0
while attempt < 3:
    attempt += 1
    if attempt == 2:
        print("PIN Accepted")
        break
    print("Wrong PIN")
```

```python
#Inventory Check
stock = 10
while stock > 0:
    stock -= 3
    if stock < 4:
        print("Low Stock")
        break
    print("Stock OK")
```

```python
#Salary Deduction
salary = 50000
deduction = 0
while salary > 40000:
    salary -= 3000
    deduction += 1
    if deduction == 2:
        break
print(salary)
```

```python
#Battery Drain
battery = 100
while battery > 0:
    battery -= 30
    if battery <= 20:
        print("Low Battery")
        break
```

```python
#Ticket Counter
tickets = 7
while tickets > 0:
    tickets -= 1
    if tickets == 2:
        print("Counter Closed")
        break
    print(tickets)
```

```python
#Data Usage Monitor
data = 0
while data < 10:
    data += 3
    if data > 8:
        print("Limit Exceeded")
        break
```

```python
#Password Strength
length = 0
while length < 8:
    length += 2
    if length == 6:
        print("Weak Password")
        break
```

```python
#Cab Distance Check
distance = 0
while distance <= 20:
    distance += 7
    if distance >= 15:
        print("High Fare")
        break
```

```python
#Exam Reattempt
attempt = 1
while attempt <= 3:
    if attempt == 3:
        print("Final Attempt")
        break
    print("Retry")
    attempt += 1
```
