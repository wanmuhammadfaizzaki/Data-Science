```python
#classify the cars
#<10000 then level1
#<25000 then level2
#<4000 then level3
#>4000 then level4
#create a new column price_status

# for i in df["price"]:
#   if i <10000:
#     print("level1")
#   elif i<25000:
#     print("level2")
#   elif i<=40000:
#     print("level3")
#   else:
#     print("level4")


def pricex(a):
  if a<10000:
    return "Level-1"
  elif a<=25000:
    return "Level-2"
  elif a<=40000:
    return "Level-3"
  else:
    return "Level-4"

df["Price_status"]=df["price"].map(pricex)

#df["Price_status"]= --> it will create the new column price_status
#df["price"].map(pricex) -> map function will map each and every value of price
#with the function of pricex

df.head()
```

```python
# Add a new column called car_category based on the following logic:

# If price is greater than or equal to 800000, label it as "Premium"

# If price is between 400000 and 799999, label it as "Mid-Range"

# If price is less than 400000, label it as "Budget"

# Use the existing price column to create this new column.
```

```python
def car_cat(b):
  if b >= 40000:
    return "Premium"
  elif b >= 20000:
    return "Mid-Range"
  else:
    return "Budget"

df["car_category"]=df["price"].map(car_cat)

```

```python
set(df["car_category"]) #set => remove duplicate
```

```python
#Grouping
#it will group the columns

#group?
#boys => 4 (count) | 28 (Avg_Age)
#girls => 4 (count) | 25 (Avg_Age)

#aggregated functions => i will aggregate all the input data
#min / max / sum / count / average


print(df["price"].max())
print(df["price"].min())
print(df["price"].mean())
print(df["price"].count())
```

```python
#analyze my data and give Top 3 years that performed well

#in revenue (total_revenue - sum)
#in car count wise (total_car_count - count)
#in average price (average)
```

```python
df2 = df.groupby(["year"]).agg(Total_revenue=("price","sum"), Total_car_count=("make","count"), Average_price=("price","mean"))
df2.sort_values("Total_revenue", ascending= False).head(3).reset_index()[["year","Average_price"]]
```

```python
df1=df.groupby(["year"]).agg(Total_revenue=("price", "sum"),Total_car_count=("make","count"), Average_price=("price","mean"))
df2=df1.sort_values("Total_car_count",ascending=False).reset_index()
df2[["year","Total_car_count"]].head(3)
```

```python
df1=df.groupby(["year"]).agg(Total_revenue=("price", "sum"),Total_car_count=("make","count"), Average_price=("price","mean"))
df2=df1.sort_values("Average_price",ascending=False).reset_index()
df2[["year","Average_price"]].head(3)
```
