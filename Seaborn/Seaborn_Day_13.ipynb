```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Q_year"]=pd.to_datetime(df["Order Date"]).dt.to_period('Q')
df = df.sort_values("Order Date")
df["Q_year"] = df["Q_year"].astype(str)

plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Q_year", y="Sales", ci = None,color ="red",marker="o",hue="Segment")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="Q_year", y="Sales",ci=None,color="red",marker="o",hue = "Segment",estimator='sum')
#estimator is used to mention aggregated function like (sum,mean,median,max,min)
#ci - Confidence Interval
plt.xticks(rotation=90)
plt.show()

#default it always take average
```

```python
import requests
import pandas as pd
import pprint

url = "https://storage.dosm.gov.my/population/population_malaysia.csv"
df = pd.read_csv(url)
```

```python
df.columns
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["date"] = pd.to_datetime(df["date"])
df["year"]=pd.to_datetime(df["date"]).dt.year
df = df.sort_values("date")
df["year"] = df["year"].astype(str)

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="year", y="population", ci = None,color ="red",marker="o",hue="sex")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="year", y="population",ci=None,color="red",marker="o",hue = "ethnicity",estimator='max')
#estimator is used to mention aggregated function like (sum,mean,median,max,min)
#ci - Confidence Interval
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df,x="age",hue="sex")
plt.xticks(rotation=90)
```
