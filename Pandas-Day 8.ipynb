```python
def sum_factorial(x,y,z):
  fx = 1
  for i in range(1,x+1):
    fx *= i
    fy =1
    for i in range(1,y+1):
      fy *= i
      fz = 1
      for i in range(1,z+1):
        fz *= i
  return fx+fy+fz
```

```python
# Do employees generally enroll after attending the webinar? Prove using dates.
df1 = pd.merge(webinar_df,employee_enrolled_df,left_on = "webinar_id",right_on = "webinar_id",how = "right")
w_d = df1["webinar_date"]
def enroll_after_attending(a):
  if a >= w_d:
    return True
  else:
    return False

df1["enroll_after_attending"] = df1["enrolled_date"].apply(enroll_after_attending)
df1[["enrolled_date","enroll_after_attending"]]
```

```python
# Compare Kuala Lumpur vs Selangor in terms of:

# Enrollment count
# Attendance rate
df1 = pd.merge(webinar_df,employee_enrolled_df,left_on = "webinar_id",right_on = "webinar_id",how = "right")
df2 = df1[(df1["city"] == "Kuala Lumpur") | (df1["city"] == "Selangor")]

df3 = df2.groupby("city").agg(Enrollment_count = ("emp_name","count"), Attendance = ("attended","count")).reset_index()
df3["Percentage"] = round(df3["Attendance"]/df3["Enrollment_count"]*100,2)
df3
```

```python
# Which city shows the highest post-webinar conversion?
df1 = pd.merge(webinar_df,employee_enrolled_df,left_on = "webinar_id",right_on = "webinar_id",how = "right")
def pw_con(a):
  count = 0
  if a == "Yes":
    count += 1
    return count
  else:
    return 0

df1["total_enrolled"] = df1["attended"].apply(pw_con)
df1.groupby("city").agg(total_enrolled = ("total_enrolled","sum")).reset_index().sort_values("total_enrolled",ascending = False).head()
```
