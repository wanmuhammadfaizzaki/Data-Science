```python
import pandas as pd

# Personal DataFrame
personal_df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 10],
    "name": ["abi", "kumar", "sheela", "mala", "kavitha", "suraj"],
    "age": [23, 34, 43, 54, 78, 34],
    "city_id": [101, 101, 102, 103, 104, 109]
})

# Salary DataFrame
salary_df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "salary": [10000, 20000, 30000, 40000, 50000, 60000, 70000, 50000, 40000]
})

print("Personal DataFrame:")
print(personal_df)

print("\nSalary DataFrame:")
print(salary_df)
```

```python
personal_df
```

```python
salary_df
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id", right_on = "id",how = "inner")

#wxact match between 2 tables
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "id",how="left")

#exact match between 2 table
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "id",how="right")

#exact match between 2 table
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "id",how="outer")

#exact match between 2 table
```

```python
import pandas as pd

# Personal DataFrame
personal_df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 10],
    "name": ["abi", "kumar", "sheela", "mala", "kavitha", "suraj"],
    "age": [23, 34, 43, 54, 78, 34],
    "city_id": [101, 101, 102, 103, 104, 109]
})

# Salary DataFrame
salary_df = pd.DataFrame({
    "Emp_id": [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "salary": [10000, 20000, 30000, 40000, 50000, 60000, 70000, 50000, 40000]
})

print("Personal DataFrame:")
print(personal_df)

print("\nSalary DataFrame:")
print(salary_df)
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "Emp_id",how="inner")

#exact match between 2 table
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "Emp_id",how="left")

#exact match between 2 table
```

```python
#inner join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "Emp_id",how="right")

#exact match between 2 table
```

```python
#outer join
pd.merge(personal_df,salary_df,left_on = "id" ,right_on = "Emp_id",how="outer")

#exact match between 2 table
```

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]

import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the file path (CSV inside the dataset)
file_path = "HRDataset_v14.csv"

# Load the latest version of the dataset
df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "rhuebner/human-resources-data-set",
    file_path
)

print("First 5 records:")
print(df.head())
```

```python
#which source people recruitmenthires - more hire
```

```python
df1 = df.groupby(["RecruitmentSource"]).agg(Total_by_sources=("RecruitmentSource","count"))
df1["rank"] = df1["Total_by_sources"].rank(ascending=False)
df1.sort_values("rank", ascending=True).reset_index().head(3)
```

```python
df1=df.groupby(["RecruitmentSource"]).agg(Total_Recruit_Source=("RecruitmentSource","count"))
df1.sort_values('Total_Recruit_Source', ascending=False).reset_index()

df1["rank"] = df1["Total_Recruit_Source"].rank(ascending=False) #descending order
df1[df1["rank"]<=5].sort_values('rank',ascending=True)
```

```python
#which source (RecruitmentSource) - more people hired
df1 = df.groupby("RecruitmentSource").agg(Total_count=("RecruitmentSource","count")).reset_index()
df1["Percentage"] = round(df1["Total_count"]/len(df)*100,2)
df1.sort_values("Total_count",ascending=False)

#inference => Iam choosing the indeed subscription it is 28%
```
