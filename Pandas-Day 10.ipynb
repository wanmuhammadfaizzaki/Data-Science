```python
df = df[df["study_hours_per_day"]!=0]
```

```python
df[["study_hours_per_day","exam_score"]].describe()
```

```python
df["study_score"] = df["exam_score"] / df["study_hours_per_day"]
```

```python
df["study_score"].describe()
```

```python
#> 20
#15 to 20
#<15

df["study_efficiency_level "]=df["study_score"].apply(lambda x : "High" if x > 20 else "Moderate" if x >=15 else "Low")
```

```python
dfm = df.groupby("study_efficiency_level ").agg(Avg_exam_score=("exam_score","mean"), Avg_time_study=("study_hours_per_day","mean")).reset_index()
dfm
```

```python
dfm["score_per_hour"]= dfm["Avg_exam_score"]/dfm["Avg_time_study"]
dfm.sort_values("score_per_hour", ascending = False)
```
