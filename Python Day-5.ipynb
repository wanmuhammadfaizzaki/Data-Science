```python
# Question 1: A fitness tracker records weight in pounds and height in inches.
#Write a function that takes these inputs, converts them to kg and meters respectively,
#calculates BMI, and classifies it as 'Underweight', 'Normal', 'Overweight', or 'Obese'.
# Write your code below this line
```

```python
name = input("Enter your name: ")
w = float(input("Enter your weight in pound: "))
h = float(input("Enter your height in inches: "))

# convert pound to kg
w = w * 0.4536
# convert inches to meters
h = h * 0.0254
#bmi formula
bmi = round((w/(h*h)),2)

if bmi < 18.5:
  result = "Underweight"
elif bmi <= 24.9:
  result ="Normal"
elif bmi <= 29.9:
  result = "Overweight"
else:
  result = "Obese"

print(f"Hi {name}!!\nyour bmi is {bmi} and classified as {result}")
```

```python
# Question 3: You have a list of dictionaries, each representing a product with 'name' and 'price'.
#Use `filter()` to get a list of products whose price is above 100 and name starts with 'A'.
# Write your code below this line
```

```python
products = [
    {"name": "Apple", "price": 150},
    {"name": "Avocado", "price": 120},
    {"name": "Banana", "price": 80},
    {"name": "Apricot", "price": 90},
    {"name": "Almond", "price": 90}
]

lt = lambda p: p["price"] > 100 and p["name"][0]=="A"

result = list(
    filter(lt, products)
)

print(result)


```

```python
# Question 4: Create a higher-order function `apply_operation(operation, x, y)` that takes a function (e.g., add, subtract) and two numbers,
#then applies the function to the numbers. Implement add and subtract functions and demonstrate.
# Write your code below this line
```

```python
def apply_operation(operation, x, y):
    return operation(x, y)

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

print(apply_operation(add, 10, 5)) # => add(10,5) => 15

print(apply_operation(subtract, 10, 5)) # => subtract(10,5) => 5
```

```python
# Question 5: Implement a function `rotate_list(lst, k)`
#that rotates a list `lst` to the right by `k` steps.
#For example, `[1,2,3,4,5]` rotated by 2 becomes `[4,5,1,2,3]`.
# Write your code below this line
```

```python
def rotate_list(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_list([1, 2, 3, 4, 5], 2))
```

```python
# Question 6: Write a function `sort_by_second(list_of_tuples)` that takes a list of tuples,
#where each tuple has two elements, and sorts the list based on the second element of each tuple.
# Write your code below this line
```

```python
def sort_by_second(data):
  return sorted(data, key = lambda x: x[1])

data = [(2,1),(3,4),(5,2),(6,7),(9,3)]
sorted_data = sort_by_second(data)
print(f"Original list: {data}")
print(f"Sorted list: {sorted_data}")
```

```python
# Question 7: You have a list of words. Find all words that appear in at
#least two different sentences (represented as sets of words).
# Write your code below this line
```

```python
sentences = [
    {"data", "science", "python"},
    {"python", "analysis", "data"},
    {"machine", "learning", "python"}
]

word_count = {}

for sentence in sentences:
    for word in sentence:
        word_count[word] = word_count.get(word, 0) + 1

result = {word for word, count in word_count.items() if count >= 2}

print(result)
```

```python
# Question 8: Given a dictionary where keys are names and values are lists of items they like.
#Write a function to find out which items are liked by all people.
# Write your code below this line
```
