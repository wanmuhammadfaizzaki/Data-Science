```python
#Q1
df1 = df.groupby("State").agg(Total_sales=("Sales","sum"), Avg_profit=("Profit","mean") ).reset_index()
df1[df1["Avg_profit"] < 0].sort_values("Total_sales", ascending=False).head()
```

```python
#Q2
df1 = df.groupby("Category").agg(total_sales=("Sales","sum"), total_profit=("Profit","sum")).reset_index()
df1["profit_margin"] = df1["total_profit"] / df1["total_sales"]
df1[df1["profit_margin"] < 0.1].sort_values("profit_margin")
```

```python
#Q3
df1 = df.groupby(["Region","Sub-Category"]).agg(Total_Profit=("Profit", "sum")).reset_index()
df2 = df1.sort_values(["Region","Total_Profit"],ascending=[True,False])
df2.groupby("Region").head(3)
```

```python
#Q5
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["year"]=df["Order Date"].dt.year
df1= df.groupby("year").agg(total_sales=("Sales","sum"), total_profit=("Profit","sum")).reset_index()
df1
```

```python
#Q6
df1  = df.groupby("City").agg(total_sales=("Sales","sum"), total_profit=("Profit","sum")).reset_index()

s_75 = df1["total_sales"].quantile(0.75)
df1[(df1["total_sales"] > s_75) & (df1["total_profit"]<0)].sort_values("total_sales")
```

```python
#Q7
df1=df[df["Discount"] > 0.4]
df2 = df1.groupby(["Category","Sub-Category"]).agg(total_orders=("Order ID","count"), T_discount=("Discount","sum")).reset_index()
df2["rate_proxy"] = (df2["T_discount"] / df2["total_orders"])*100
df2[df2["rate_proxy"]>= 30].sort_values("rate_proxy",ascending=False).head(10)
```

```python
#Q8
df1 = df.groupby("Customer Name").agg(Order_Count = ("Order ID","count"),Avg_profit=("Profit","mean")).reset_index()
df1[(df1["Order_Count"]>5) & (df1["Avg_profit"]<0)].sort_values("Avg_profit")
```

```python
#Q9
df["Delay"] = df["Ship Date"]-df["Order Date"]
df[df["Delay"]> "5 days"].sort_values("Delay", ascending= False)[["Ship Date","Order Date","Delay"]]
```

```python
#Q10
df1 = df.groupby("State").agg(total_sale=("Sales","sum"), total_profit=("Profit","sum"), no_of_orders=("Order ID","count")).reset_index()
df1.describe()
```

```python
sales_med = df1["total_sale"].median()
profit_med = df1["total_profit"].median()
def category(x,y):
  if x > sales_med and y > profit_med:
    return "High Sales – High Profit"
  elif x > sales_med and y < profit_med:
    return "High Sales – Low Profit"
  elif x < sales_med and y > profit_med:
    return "Low Sales – High Profit"
  else:
    return "Low Sales – Low Profit"
df1["Category"] = df1.apply(lambda x: category(x["total_sale"], x["total_profit"]),axis=1)
df1
```
