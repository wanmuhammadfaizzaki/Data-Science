```python
import pandas as pd

sheet_id = "1RVnUAGVsr4lrt7nZ2WOq2hFkj7gFcRef"
sheet_name = "Orders"  # or the visible tab name

url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
df = pd.read_csv(url)

df
```

```python
import numpy as np
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["sales_period"] = np.where(df["Order Date"] < "2015-01-01", "Before 2015", "After 2015")
df["Order_date_month"] = df["Order Date"].dt.to_period("M")
df1 = df.groupby(["sales_period","Order_date_month"]).agg(avg_sales=("Sales","mean")).reset_index()
df1 = df1[df1["sales_period"] == "Before 2015"]
df1["Order_date_month_float"] = (df1["Order_date_month"].dt.year +(df1["Order_date_month"].dt.month - 1) / 12)
df1
```

```python
df2 = df.groupby(["sales_period","Order_date_month"]).agg(avg_sales=("Sales","mean")).reset_index()
df2 = df2[df2["sales_period"] == "After 2015"]
df2["Order_date_month_float"] = (df2["Order_date_month"].dt.year +(df2["Order_date_month"].dt.month - 1) / 12)
df2
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))


line_red, = plt.plot(
    df1["Order_date_month_float"],
    df1["avg_sales"],
    linewidth=1,
    linestyle="--",
    marker="o",
    markersize=8,
    markerfacecolor="red",
    markeredgewidth=2,
    color="red",
    label="Before 2015"  # Best practice: add label here
)

line_blue, = plt.plot(
    df2["Order_date_month_float"],
    df2["avg_sales"],
    linewidth=1,
    linestyle="--",
    marker="+",
    markersize=8,
    markerfacecolor="blue",
    markeredgewidth=2,
    color="blue",
    label="After 2015"   # Best practice: add label here
)

plt.title(
    "Average Sales per Month",
    fontsize=18,
    fontweight="bold",
    pad=10
)

plt.xlabel("Year", fontsize=13, labelpad=10)
plt.ylabel("Average Sales", fontsize=13, labelpad=10)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)


plt.legend(handles=[line_red, line_blue], loc="lower right", title="Legend")

plt.grid(
    True,
    linestyle=":",
    linewidth=0.8,
    alpha=0.8
)

plt.show()
```
