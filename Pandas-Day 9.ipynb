```python
def loss(x):
  if x < 0:
    return x
  else:
    return 0

df["loss"] = df["Profit"].apply(loss)
df1 = df.groupby("Category").agg(Total_loss = ("loss", "sum")).reset_index()
df1["rank"] = df1["Total_loss"].rank(ascending= True)
df1[["rank","Category","Total_loss"]]
```

```python
df2 = df.groupby("Sub-Category").agg(Total_loss = ("loss", "sum")).reset_index()
df2["rank"] = df2["Total_loss"].rank(ascending= True)
df2.sort_values("Total_loss").head(3)

```

```python
df[df["Category"] == "Furniture"].groupby("Sub-Category").agg(Total_loss = ("loss", "sum")).reset_index().sort_values("Total_loss", ascending= True)
```

```python
df1 = df[(df["Category"] == "Office Supplies") ].groupby("Sub-Category").agg(Total_loss = ("loss", "sum")).reset_index().sort_values("Total_loss", ascending= True)
df1[df1["Total_loss"]<0]
```

```python
df2 = df[(df["Category"] == "Technology") ].groupby("Sub-Category").agg(Total_loss = ("loss", "sum")).reset_index().sort_values("Total_loss", ascending= True)
df2[df2["Total_loss"]<0]
```

```python
df3 = df.groupby("Segment").agg(Total_loss = ("loss", "sum")).reset_index()
df3["rank"] = round(df3["Total_loss"].rank(ascending= True))
df3.sort_values("Total_loss").head(3)
```

```python
df4 = df.groupby("Product Name").agg(Total_loss = ("loss", "sum")).reset_index()
df4["rank"] = round(df4["Total_loss"].rank(ascending= True))
df4.sort_values("Total_loss").head(3)
```

📌 Shipping & Delivery Optimization

Operations team wants to reduce logistics cost.

Question: Analyze shipping modes vs delivery time vs profit. Which shipping mode results in higher delays or lower profit, and how should the company optimize shipping strategy?

```python
#delivery_time = shipp_date - order_date
```

```python
df.dtypes
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y")
df["Delivery_Time"] = df["Ship Date"] - df["Order Date"]
df1= df.groupby("Ship Mode").agg(Average_Delay = ("Delivery_Time","mean"),Average_profit = ("Profit","mean")).reset_index().sort_values("Average_Delay", ascending = False)
df1["Average_Delay"] = df1["Average_Delay"].dt.days
df1[(df1["Average_Delay"]>=3) | (df1["Average_profit"] < 28)]
```
